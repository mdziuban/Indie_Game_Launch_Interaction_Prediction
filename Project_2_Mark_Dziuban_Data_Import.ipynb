### Data Import Notebook

```python
import pandas as pd
```

#### Organizing the data

First I will read the csv data files and store them into dataframes.  From there, they will be merged into one dataframe on the column appid.  

```python
steam_app_df = pd.read_csv('steam_app_data.csv')
steamspy_df = pd.read_csv('steamspy_data.csv')
```

```python
steam_app_df.head()
```

```python
steamspy_df.head()
```

```python
# I want to merge the two dataframes on the appid column.  In the steam_app_df the column is labeled steam_appid 
# and must be renamed.

steam_app_df = steam_app_df.rename(columns={"steam_appid":"appid"})
```

```python
merged_steam_df = pd.merge(steamspy_df,steam_app_df, on="appid")
```

```python
merged_steam_df.columns
```

```python
merged_steam_df.info()
```

```python
# Store the raw data in a csv for easy access if there is a problem later.  

merged_steam_df.to_csv('raw_merged_steam_df.csv')
```
